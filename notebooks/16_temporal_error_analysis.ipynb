```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Paths
PREDICTIONS_DIR = Path("../predictions")

# Assets
ASSETS = ["AAPL", "IBEX35", "MSFT", "NASDAQ", "S&P500"]

# Matplotlib style
plt.style.use("seaborn-v0_8-darkgrid")
```

```python
# Error calculation function
def compute_errors(df):
    df = df.copy()
    df["error"] = df["y_true"] - df["y_pred"]       # signed error
    df["abs_error"] = df["error"].abs()             # absolute error
    df["squared_error"] = df["error"] ** 2          # squared error
    return df
```

```python
for asset in ASSETS:
    print(f"\n=== Temporal Error Analysis for {asset} ===")
    
    # Load predictions
    lstm_df = pd.read_csv(PREDICTIONS_DIR / f"predictions_lstm_{asset}.csv", parse_dates=["date"])
    tft_df = pd.read_csv(PREDICTIONS_DIR / f"predictions_tft_{asset}.csv", parse_dates=["date"])
    
    # Compute errors
    lstm_df = compute_errors(lstm_df)
    tft_df = compute_errors(tft_df)
    
    # Plo1: Signed error over time
    plt.figure(figsize=(12, 5))
    plt.plot(lstm_df["date"], lstm_df["error"], label="LSTM Error", color="blue", alpha=0.6)
    plt.plot(tft_df["date"], tft_df["error"], label="TFT Error", color="red", alpha=0.6)
    plt.axhline(0, color="black", linestyle="--", linewidth=1)
    plt.title(f"Signed Prediction Error Over Time - {asset}")
    plt.xlabel("Date")
    plt.ylabel("Error (y_true - y_pred)")
    plt.legend()
    plt.show()
    
    # Plot 2: Rolling MAE over time
    window = 20  # 20 trading days (~1 month)
    lstm_df["rolling_mae"] = lstm_df["abs_error"].rolling(window).mean()
    tft_df["rolling_mae"] = tft_df["abs_error"].rolling(window).mean()
    
    plt.figure(figsize=(12, 5))
    plt.plot(lstm_df["date"], lstm_df["rolling_mae"], label=f"LSTM Rolling MAE ({window}d)", color="blue")
    plt.plot(tft_df["date"], tft_df["rolling_mae"], label=f"TFT Rolling MAE ({window}d)", color="red")
    plt.title(f"Rolling Mean Absolute Error - {asset}")
    plt.xlabel("Date")
    plt.ylabel("Rolling MAE")
    plt.legend()
    plt.show()
    
    # Plot 3: Absolute error vs true volatility
    plt.figure(figsize=(7, 6))
    plt.scatter(lstm_df["y_true"], lstm_df["abs_error"], alpha=0.5, label="LSTM", color="blue")
    plt.scatter(tft_df["y_true"], tft_df["abs_error"], alpha=0.5, label="TFT", color="red")
    plt.xlabel("True Volatility")
    plt.ylabel("Absolute Error")
    plt.title(f"Error vs True Volatility - {asset}")
    plt.legend()
    plt.show()
```

```python
# Save extended error datasets for future analysis
for asset in ASSETS:
    lstm_df = pd.read_csv(PREDICTIONS_DIR / f"predictions_lstm_{asset}.csv", parse_dates=["date"])
    tft_df = pd.read_csv(PREDICTIONS_DIR / f"predictions_tft_{asset}.csv", parse_dates=["date"])
    
    lstm_df = compute_errors(lstm_df)
    tft_df = compute_errors(tft_df)
    
    lstm_df.to_csv(PREDICTIONS_DIR / f"errors_lstm_{asset}.csv", index=False)
    tft_df.to_csv(PREDICTIONS_DIR / f"errors_tft_{asset}.csv", index=False)

print("Error analysis completed and saved.")
```
