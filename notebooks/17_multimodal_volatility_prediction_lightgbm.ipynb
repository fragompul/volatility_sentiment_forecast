```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler

import lightgbm as lgb
```

```python
# Paths
DATA_DIR = Path("../data/processed")
OUTPUT_MERGED = Path("../data/processed/merged_all_assets.csv")
MODELS_DIR = Path("../models")
MODELS_DIR.mkdir(exist_ok=True)

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
```

```python
# Load sentiment data (normalize Date column)
sentiment_df = pd.read_csv(DATA_DIR / "sentiment_features.csv")
# Rename 'date' to 'Date' in sentiment data to unify with market datasets
sentiment_df.rename(columns={"date": "Date"}, inplace=True)
sentiment_df["Date"] = pd.to_datetime(sentiment_df["Date"])
sentiment_df = sentiment_df.sort_values("Date")

# Initialize list for all merged datasets
all_assets_df = []

for feature_file in DATA_DIR.glob("*_features.csv"):
    if "sentiment" in feature_file.name:
        continue  # skip sentiment file

    asset_name = feature_file.stem.replace("_features", "")
    market_df = pd.read_csv(feature_file)
    market_df["Date"] = pd.to_datetime(market_df["Date"])
    market_df = market_df.sort_values("Date")

    # Merge keeping all market dates
    merged_df = pd.merge_asof(
        market_df,
        sentiment_df,
        on="Date",
        direction="backward",   # take last known sentiment before/at date
        allow_exact_matches=True
    )

    merged_df["Ticker"] = asset_name

    # Fill missing sentiment data
    sentiment_cols = [
        "tweet_sentiment_mean", "tweet_sentiment_std", "tweet_count",
        "news_sentiment_mean", "news_sentiment_std", "news_count"
    ]
    merged_df[sentiment_cols] = merged_df[sentiment_cols].fillna(method="ffill").fillna(0)

    all_assets_df.append(merged_df)

# Combine all assets
df_all = pd.concat(all_assets_df, ignore_index=True)
df_all = df_all.sort_values(["Ticker", "Date"]).reset_index(drop=True)

print(f"Final dataset shape after merge and fill: {df_all.shape}")
df_all.head()
```

```python
# Save dataset
df_all.to_csv(OUTPUT_MERGED, index=False)
print(f"Saved merged dataset: {df_all.shape} rows")
```

```python
# Target variable: short-term realized volatility
TARGET = "Volatility_5d"

# Features: market + sentiment
feature_cols = [
    # Market features
    "Return", "Volatility_21d", "RSI_14", "MACD", "MACD_signal",
    "MACD_hist", "ATR_14", "SMA_20", "SMA_50", "SMA_200",
    # Sentiment features
    "tweet_sentiment_mean", "tweet_sentiment_std", "tweet_count",
    "news_sentiment_mean", "news_sentiment_std", "news_count"
]

# Drop rows with NaNs in key columns
df_all = df_all.dropna(subset=feature_cols + [TARGET])

print(f"Final dataset shape after cleaning: {df_all.shape}")
```

```python
# Ensure Date column is datetime
df_all["Date"] = pd.to_datetime(df_all["Date"])

# Sort by date to avoid leakage
df_all = df_all.sort_values("Date").reset_index(drop=True)

# Train/test split by index (80% train, 20% test)
split_idx = int(len(df_all) * 0.8)
train_df = df_all.iloc[:split_idx]
test_df  = df_all.iloc[split_idx:]

print(f"Train size: {len(train_df)}, Test size: {len(test_df)}")

# Features and target
X_train = train_df[feature_cols]
y_train = train_df[TARGET]
X_test  = test_df[feature_cols]
y_test  = test_df[TARGET]
```

```python
# Scaling features
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled  = scaler_X.transform(X_test)

scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).ravel()
y_test_scaled  = scaler_y.transform(y_test.values.reshape(-1, 1)).ravel()

# Save scalers
joblib.dump(scaler_X, MODELS_DIR / "scaler_X_multimodal.pkl")
joblib.dump(scaler_y, MODELS_DIR / "scaler_y_multimodal.pkl")
```

```python
# LightGBM parameters
params = {
    "objective": "regression",
    "metric": "rmse",
    "boosting_type": "gbdt",
    "num_leaves": 31,
    "learning_rate": 0.05,
    "feature_fraction": 0.9,
    "bagging_fraction": 0.8,
    "bagging_freq": 5,
    "seed": RANDOM_SEED
}

# Create dataset for LightGBM
lgb_train = lgb.Dataset(X_train_scaled, y_train_scaled)
lgb_test = lgb.Dataset(X_test_scaled, y_test_scaled, reference=lgb_train)

# Train model
model = lgb.train(
    params,
    lgb_train,
    valid_sets=[lgb_train, lgb_test],
    num_boost_round=2000,
    callbacks=[
        lgb.early_stopping(50),
        lgb.log_evaluation(100)
    ]
)

# Save model
model.save_model(str(MODELS_DIR / "volatility_model_multimodal_lgb.txt"))
```

```python
# Predict (scaled space)
y_pred_scaled = model.predict(X_test_scaled)

# Inverse scaling
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).ravel()
y_true = scaler_y.inverse_transform(y_test_scaled.reshape(-1, 1)).ravel()
```

```python
# Metrics
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mae = mean_absolute_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)

print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RÂ²: {r2:.4f}")
```

```python
# Save predictions
pred_df = pd.DataFrame({
    "Date": test_df["Date"].values,
    "Ticker": test_df["Ticker"].values,
    "Actual_Volatility": y_true,
    "Predicted_Volatility": y_pred
})

pred_df.to_csv(MODELS_DIR / "predictions_multimodal_lgb.csv", index=False)
pred_df.head()
```

```python
# Plot feature importance
lgb.plot_importance(model, max_num_features=15, importance_type="gain")
plt.title("Feature Importance - LightGBM Multimodal Model")
plt.show()
```

```python
# Proper feature importance plot with feature names
importances = model.feature_importance(importance_type="gain")
feature_names = feature_cols  # keep your original column names
importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(importance_df["Feature"], importance_df["Importance"])
plt.gca().invert_yaxis()
plt.title("Feature Importance - LightGBM Multimodal Model")
plt.xlabel("Gain Importance")
plt.show()
```

```python
# Visualization of predictions
ASSETS = ["AAPL", "IBEX35", "MSFT", "NASDAQ", "S&P500"]

for asset in ASSETS:
    asset_df = pred_df[pred_df["Ticker"] == asset]

    plt.figure(figsize=(12,5))
    plt.plot(asset_df["Date"], asset_df["Actual_Volatility"], label="Actual")
    plt.plot(asset_df["Date"], asset_df["Predicted_Volatility"], label="Predicted")
    plt.title(f"Multimodal Volatility Prediction - {asset}")
    plt.xlabel("Date")
    plt.ylabel("Volatility")
    plt.legend()
    plt.show()
```
