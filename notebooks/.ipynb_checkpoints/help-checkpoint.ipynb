```python
import os

# Diccionario con los nombres de carpetas y sus rutas relativas desde 'notebooks'
folders = {
    'notebooks': '.',
    'models': '../models',
    'predictions': '../predictions',
    'data/raw': '../data/raw',
    'data/processed': '../data/processed'
}

# Recorremos cada carpeta y listamos los archivos que contiene
folder_files = {}
for name, path in folders.items():
    try:
        files = [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]
        folder_files[name] = files
    except FileNotFoundError:
        folder_files[name] = ['[Ruta no encontrada]']

# Imprime el resultado (opcional)
for folder, files in folder_files.items():
    print(f"{folder}:")
    for file in files:
        print(f"  - {file}")
    print()

# Si quieres usar el diccionario en tu c√≥digo
# puedes accederlo como 'folder_files'
```
